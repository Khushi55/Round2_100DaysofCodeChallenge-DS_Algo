```python
from datetime import datetime

def Bubblesort(arr):
    tstart = datetime.now()  
    n = len(arr) 
  
    
    for i in range(n): 
        for j in range(0, n-i-1):  
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
    tend = datetime.now()
    t=tend-tstart
    
    return (t.total_seconds())             
```

```python

f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0
for p in range(0,3):
    f=s[p]
    
    
    for i in range(1,11):
        A=[]
        j=0
        f.seek(0)
        for line in f:
            if j==i*10000:
                break
            A.append(int(line))
        
        
            j+=1  
        k=Bubblesort(A)    
        K[x]=k
        x+=1
        
        print(i)
    
print(K)


```

```python
import matplotlib.pyplot as plt
X=[10000,20000,30000,40000,50000,60000,70000,80000,90000,100000]
Y1=K[0:10]

Y2=K[10:20]
Y3=K[20:30]
plt.figure(figsize=(8,6))
plt.plot(X,Y1,label="Average Case")

plt.plot(X,Y2,label="Worst Case")
plt.plot(X,Y3,label="Best Case")
plt.xlabel("x-axis") # naming the x axis
plt.ylabel("y-axis")  # naming the y axis
plt.title("Bubble Sort") # title of the graph
plt.legend()
plt.show()
```

```python
def insertion_Sort(A):
    
    for i in range(1,len(A)):
        key=A[i]
        j=i-1
        while j>=0 and key<A[j]:
            A[j+1]=A[j]
            j-=1
        A[j+1]=key
    
    
```

```python
import time
f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0
for p in range(0,3):
    f=s[p]
    for i in range(1,11):
        A=[]
        j=0
        f.seek(0)
        for line in f:
            if j==i*10000:
                break
            A.append(int(line))
        
        
            j+=1 
        tstart = time.clock()   
        insertion_Sort(A)
        tend = time.clock()
        t=tend-tstart
    
        k=t*1000
        K[x]=k
        x+=1
        print(i)
print(K)

```

```python
def selection_sort(A):
    for i in range(len(A)):
        minindex=i
        for j in range(i+1,len(A)):
            if A[minindex]>A[j]:
                minindex=j
        A[i],A[minindex]=A[minindex],A[i]
   
```

```python
import time
f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0

for p in range(0,3):
    f=s[p]
    A=[]
    f.seek(0)
    
    for line in f:
        A.append(int(line))
    for i in range(1,11):
        
        
        
        tstart = time.clock()   
        selection_sort(A[0:i*10000])
        tend = time.clock()
        t=(tend-tstart)
    
        k=t
        K[x]=k
        x+=1
        print(i)
print(K)
```

```python
class HeapSort:
    def __init__(self,alist):
        self.alist=[i for i in alist]
        
        self.length=len(self.alist)
        if self.length<1:
            return
        self.buildHeap()
       
        
        while self.length>=1:
            max=self.Extract_max(self.length)
            
            self.length-=1
            self.alist[self.length]=max
        
        
        
        
    def parent(self,i):
        return (i-1)//2
    def left_child(self,i):
        
        return 2*i +1
    def right_child(self,i):
        return 2*i+2
    def isleft(self,i):
        return (self.left_child(i)<len(alist))
    def isright(self,i):
        return (self.right_child(i)<len(alist))

    def buildHeap(self):
        if len(self.alist)<=0:
            return 
        start=self.parent(self.length-1)
        for i in range(start,-1,-1):
            self.maxheapify(i,self.length)

    def maxheapify(self,index,size):
        if index<=-1:
            return
        left=self.left_child(index)
        if left<size and self.alist[left]>self.alist[index]:
            largest=left
        else:
            largest=index
        right = self.right_child(index)
        if right<size and self.alist[right]>self.alist[largest]:
            largest=right
       
        if largest!=index:
            self.alist[largest],self.alist[index]=self.alist[index],self.alist[largest]
            self.maxheapify(largest,size)
        
    def Extract_max(self,size):
        
        max=self.alist[0]
        self.alist[0]=self.alist[size-1]

        self.maxheapify(0,size)
        return max
                
```

```python
import time
f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0

for p in range(0,3):
    f=s[p]
    A=[]
    f.seek(0)
    
    for line in f:
        A.append(int(line))
    for i in range(1,11):
        
        
        
        tstart = time.clock()   
        HeapSort(A[0:i*10000])
        tend = time.clock()
        t=(tend-tstart)
    
        k=t
        K[x]=k
        x+=1
print(K)
```
