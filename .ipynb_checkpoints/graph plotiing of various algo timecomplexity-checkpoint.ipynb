```python
from datetime import datetime

def Bubblesort(arr):
    tstart = datetime.now()  
    n = len(arr) 
  
    
    for i in range(n): 
        for j in range(0, n-i-1):  
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
    tend = datetime.now()
    t=tend-tstart
    
    return (t.total_seconds())             
```

```python

f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0
for p in range(0,3):
    f=s[p]
    
    
    for i in range(1,11):
        A=[]
        j=0
        f.seek(0)
        for line in f:
            if j==i*10000:
                break
            A.append(int(line))
        
        
            j+=1  
        k=Bubblesort(A)    
        K[x]=k
        x+=1
        
        print(i)
    
print(K)


```

```python
def insertion_Sort(A):
    
    for i in range(1,len(A)):
        key=A[i]
        j=i-1
        while j>=0 and key<A[j]:
            A[j+1]=A[j]
            j-=1
        A[j+1]=key
    
    
```

```python
def selection_sort(A):
    for i in range(len(A)):
        minindex=i
        for j in range(i+1,len(A)):
            if A[minindex]>A[j]:
                minindex=j
        A[i],A[minindex]=A[minindex],A[i]
   
```

```python
class HeapSort:
    def __init__(self,alist):
        self.alist=[i for i in alist]
        
        self.length=len(self.alist)
        if self.length<1:
            return
        self.buildHeap()
       
        
        while self.length>=1:
            max=self.Extract_max(self.length)
            
            self.length-=1
            self.alist[self.length]=max
        
        
        
        
    def parent(self,i):
        return (i-1)//2
    def left_child(self,i):
        
        return 2*i +1
    def right_child(self,i):
        return 2*i+2
    def isleft(self,i):
        return (self.left_child(i)<len(alist))
    def isright(self,i):
        return (self.right_child(i)<len(alist))

    def buildHeap(self):
        if len(self.alist)<=0:
            return 
        start=self.parent(self.length-1)
        for i in range(start,-1,-1):
            self.maxheapify(i,self.length)

    def maxheapify(self,index,size):
        if index<=-1:
            return
        left=self.left_child(index)
        if left<size and self.alist[left]>self.alist[index]:
            largest=left
        else:
            largest=index
        right = self.right_child(index)
        if right<size and self.alist[right]>self.alist[largest]:
            largest=right
       
        if largest!=index:
            self.alist[largest],self.alist[index]=self.alist[index],self.alist[largest]
            self.maxheapify(largest,size)
        
    def Extract_max(self,size):
        
        max=self.alist[0]
        self.alist[0]=self.alist[size-1]

        self.maxheapify(0,size)
        return max
                
```

```python

def BinarySearch(Arr,ele):
    first=0
    last=len(Arr)-1
    count=0
    
    while(first<=last):
        count+=1
        m=(first+last)//2
        if Arr[m]<ele:
            first=m+1
        elif Arr[m]>ele:
            last=m-1
        else:
            print(m+1)
            break
       
    return count


            
```

```python
def LinearSearch(Arr,ele):
    length=len(Arr)
    for i in range(length):
        if Arr[i]==ele:
            return i+1
    return -1

        
```

```python
def Merge(A,p,q,r):
    k=0
    i=p
    j=q+1
    B=[None]*(r-p)
    while(i<=q and j<r):
    
        if A[i]<=A[j]:
            B[k]=A[i]
            k+=1
            i+=1
        else:
            B[k]=A[j]
            k+=1
            j+=1
    while(i<=q):
        B[k]=A[i]
        k+=1
        i+=1
    while(j<r):
        B[k]=A[j]
        k+=1
        j+=1
        
    A[p:r]=B[0:r-p+1]    
    
    
```

```python
def MergeSort(A,p,r):
    
        if(p<r):
            q=(p+r)//2
            MergeSort(A,p,q)
            MergeSort(A,q+1,r)
            Merge(A,p,q,r)
    
```

## Quick sort

```python
def partition(A,p,r):
    x=A[r]
    i=p-1
    for j in range(p,r):
        if A[j]<=x:
            i=i+1
            
          
            A[i],A[j]=A[j],A[i]
    A[i+1],A[r]=A[r],A[i+1]
    
    return i+1

        



def QuickSort(A,p,r):
    if p<r:
        q=partition(A,p,r)
        QuickSort(A,p,q-1)
        QuickSort(A,q+1,r)

arr=[85,24,63,45,17,31,96,50]
QuickSort(arr,0,len(arr)-1)        
print(arr)
```

```python
import time
f1 = open("random.txt", "rt")
f2 = open("descending.txt", "rt")
f3 = open("ascending.txt", "rt")
s=[f1,f2,f3]
K=[0]*30
x=0

for p in range(0,2):
    f=s[p]
    A=[]
    f.seek(0)
    
    for line in f:
        A.append(int(line))
    for i in range(1,11):
        
        
        
        tstart = time.clock()  
        S=A[0:i*1000]
       
        QuickSort(S,0,(i*1000)-1)
        tend = time.clock()
        
        
       
        t=(tend-tstart)
    
        k=t*1000
        K[x]=k
        x+=1
            
        
print(K)
```
